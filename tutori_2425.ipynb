```python
import pandas as pd
```

```python
tutors = pd.read_csv('~/Documenten/tutors_2425.csv')  
tutors.head()
```

Delete all the text between parenthesis

```python
tutors['Classe'] = tutors['Classe'].str.replace(r'\(.*$', '', regex=True)
tutors.head()
```

Split multiples values

```python
tutors["Classe"] =  tutors["Classe"].str.split("+")
tutors.head()
```

Eliminate double values in column classe by assigning each class to correct professor.

```python
tutors = tutors.explode("Classe")
tutors.head()
```

Eliminate white spaces

```python
tutors["Classe"] = tutors["Classe"].str.strip()
```

```python
# Corrected code to exclude the first two characters and check the rest
tutors['Indirizzo'] = tutors['Classe'].apply(lambda x: 
                                             'Classico' if 'C' in x[2:] else
                                             'Linguistico' if 'L' in x[2:] else
                                             'Scienze Umane Economico' if x[2:].endswith('SUE') else
                                             'Scienze Umane')

# Sort by 'Indirizzo' and 'Classe'
grouped_df = tutors.sort_values(by=['Indirizzo', 'Classe'], ascending=[False, True])

# Select relevant columns
grouped_df = grouped_df[["Classe", "Professore", "Indirizzo"]]

# Display the last 20 rows of the final DataFrame
grouped_df
```

```python
# Save to Excel
grouped_df.to_excel('grouped_data.xlsx', index=False)
```
